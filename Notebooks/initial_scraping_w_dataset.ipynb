```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
```

```python
#import csv file pulled from kaggle
college_df = pd.read_csv('cbb.csv')
college_df
```

```python
college_df.rename(columns = {'TEAM':'School'},inplace = True)
college_df
```

```python
#open chrome using selenium to url
driver = webdriver.Chrome()
url = "https://www.ncsasports.org/mens-basketball/division-1-colleges"
driver.get(url)
```

```python
#search for college list on website
container = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "section.wp-block-ncsa-college-list")))

rows = container.find_elements(By.CSS_SELECTOR, '[itemprop = "itemListElement"]')
```

```python
ncsa_data = []
```

```python
#use for loop to extract school info
for row in rows:
    school_element = row.find_element(By.CSS_SELECTOR, '[itemprop="name"]')
    address_element = row.find_element(By.CSS_SELECTOR, '[itemprop="address"]')

    school = school_element.text.strip()
    city_state = address_element.text.strip()

    if 'Public' in row.text:
        school_type = 'Public'
    elif 'Private' in row.text:
        school_type = 'Private'
    else:
        school_type = 'Unknown'

    row_data = {
        'School':school,
        'Location': city_state,
        'Type':school_type}
    
    ncsa_data.append(row_data)
```

```python
ncsa_data
```

```python
#convert dictionary to dataframe
df_ncsa = pd.DataFrame(ncsa_data)
df_ncsa.to_csv('ncsa_scraped.csv')
df_ncsa
```

```python
#split location into city and state columns
df_ncsa[['City','State']] = df_ncsa['Location'].str.split(',', expand = True)
df_ncsa['City'] = df_ncsa['City'].str.strip()
df_ncsa['State'] = df_ncsa['State'].str.strip()
```
